```python
## compsritive model kmeans and gmm
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv('Clustering_gmm.csv')# dataframe is made as dataset
```

```python
dataset.head(10)# there are 2 ip features every sample vector having 2 fetaures weught and heights
```

```python
plt.figure()
plt.scatter(dataset['Weight'],dataset['Height'])
plt.xlabel('Weight')## 4 clusters are formed and not use of elbow method directly 4 clusters
plt.ylabel('Height')
```

```python
kmeans=KMeans(n_clusters=4)## k means model api is formed no. of clusters are 4 in it
kmeans.fit(dataset)#3 fitting the model in the dataset + cal centroid datapoint  closer + one closest daatpoint in to that clusters
```

```python
predictions=kmeans.predict(dataset)
```

```python
dataframe1=pd.DataFrame(dataset)## copy of dataset as df1 has 2 columns + add new column predictions
```

```python
dataframe1['predictions']=predictions
dataframe1
```

```python
color=['red','yellow','green','blue']#list of colors
```

```python
for i in range(0,4):# for loop 4 times coz make 4 clusters + subsetting with i = 0,1,2,3 values + roundshaped clusters made thats why its wrong coz of k menas + a
  data=dataframe1[dataframe1['predictions']==i]
  plt.scatter(data['Weight'],data['Height'],c=color[i])
plt.show()
```

```python
dataset_gmm=pd.read_csv('Clustering_gmm.csv')
```

```python
gmm=GaussianMixture(n_components=4)## applying gmm + 4 gaussian components + dataset 4 gaussian distributions using expection + maximisation algo
```

```python
gmm.fit(dataset_gmm)# training model predicted duplicate dataset
```

```python
pred=gmm.predict(dataset_gmm)# PREDICTIONS
```

```python
dataframe2=pd.DataFrame(dataset_gmm)
```

```python
dataframe2['predictions']=pred
dataframe2
```

```python
color=['red','yellow','green','blue']# I = SAME AS ABOVE
for i in range(0,4):
  data=dataframe2[dataframe2['predictions']==i]
  plt.scatter(data['Weight'],data['Height'],c=color[i])
plt.show()
```
